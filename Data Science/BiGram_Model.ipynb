```python
# Lab Assignment: Given the text document (attached) do the following
# 1. Compute term frequency of each unique word and display the top 5 most frequent words
# 2. Compute the type token ratio of the document
# 3.  Plot the word frequency distribution and check whether the corpus is following Zipfs law 
# 4. Compute frequencies for every two consecutive words.
```

```python
import pandas as pd
```

```python
freq = {}

with open('text.txt', 'r') as file:

  for line in file:
    for word in line.split():
      if word in freq:
        freq[word] += 1
      else:
        freq[word] = 1

items = list(freq.items())

def compare(items):
  key, value = items
  return value

items.sort(key = compare, reverse = True)

sorted_freq = {}

for key, value in items:
  sorted_freq[key] = value

# print(freq)
# print(sorted_freq)

count = 0;

for key in sorted_freq:
  print("{} -> {}".format(key, sorted_freq[key]))
  count += 1;

  if (count == 5):
    break
```

```python
#type token ratio

unique_words = len(freq)

total_no_of_words = 0

for key in freq:
  total_no_of_words += freq[key]

# print(total_no_of_words)
# print(unique_words)

ratio = unique_words / total_no_of_words

print(ratio)
```

```python
import matplotlib.pyplot as plt

labels = freq.keys()
values = freq.values()

plt.plot(labels,values)
plt.xlabel('Words')
plt.ylabel('frequency')
plt.show()
```
